```python
#Name basic variables 
runFrom = 'FND4' #fnd4 or cpro2_eeg
dataType = 'RawSensor' #sensor or source
classify_cond = ['Left','Right']


#Name directories 
inputDir_2 = 'results/DynamicDecoding/RawSensor/GroupResults/'
if runFrom =='FND4':
    baseDir = '/home/let83/FND4/'
else:
    baseDir = '/projectsn/f_mc1689_1/cpro2_eeg/'

def loadData(baseDir,inputDir_2,ii):
    outputDir = f'{baseDir}{inputDir_2}{ii}/'
    inputDir = f'{baseDir}{inputDir_2}{ii}/'
    #_GroupAccuracyTseries_Left_10fold_correctOnly_noPseudoTrials.mat
    inputPrefix = f'_GroupAccuracyTseries_{ii}_10fold_correctOnly_noPseudoTrials.mat'
    datafile = f'{inputDir}{inputPrefix}'
    
    with h5py.File(datafile, 'r') as f:
        print('')
        print(list(f.keys()))  # See top-level variables
        print('')
        keys = list(f.keys())
        data = {}
        for key in keys:
            print(f'{key}')
            if isinstance(f[key],h5py.Dataset):
                data[key] = f[key][:]
                print(f'{key} added to DATA\n')
            elif key == '#refs#':
                print('Skip key!\n')
            elif isinstance(f[key],h5py.Group):
                miniDf = {}
                miniKeys = list(f[key].keys())
                for mk in miniKeys:
                    if isinstance(f[key][mk], h5py.Dataset):
                        miniDf[mk]=list(f[key][mk][:])
                        #print(f'    {key}, {mk} added to MINI\n')
                    elif isinstance(f[key][mk], h5py.Group):
                        extraMini = {}
                        miniKeys2 = list(f[key][mk].keys())
                        for mk2 in miniKeys2:
                            extraMini[mk2]=list(f[key][mk][mk2][:])
                            #print(f'      {key}, {mk}, {mk2} added to EXTRA\n')
                        miniDf[mk]=extraMini
                        #print(f'    EXTRA added to MINI ({key}, {mk})\n')
                data[key]=miniDf
                print(f'MINI added to DATA ({key})\n')
    return data

#Write function to plot left or right hand data
def plotSingleCond(group, subject, inputPrefixOutput, cond):
    #outputDir = f'{inputPrefixOutput}GroupResults/{cond}/'
    outfile = f'{inputPrefixOutput}GroupPlot_{cond}_FingerDecoding_WilcoxonBonf05'
    if cond=='Left': 
        c='steelblue'
        c1 = 'lightskyblue'
    else: 
        c='mediumorchid'
        c1='violet'
    x = np.linspace(-0.5,0.5,len(subject))
    y = group['avg']

    err = group['sterr']
    err1 = []
    err2 = []
    for i in range(len(err)):
        err1.append(y[i]-err[i])
        err2.append(y[i]+err[i])
    err1 = np.array(err1).flatten()
    err2=np.array(err2).flatten()
    timepoints = len(y)
    y_min = 0.4
    y_max = 0.9
    plt.figure()
    plt.plot(x,y,color=c,linewidth=1)
    plt.fill_between(x, err1, err2, alpha=0.3, color=c1)
    indx = 0
    for i in group['stat_wilcoxon_Bonf']['h']:
        if group['stat_wilcoxon_Bonf']['h']==1:
            plt.scatter(indx, 0.44, color=c, s=60, zorder=5)
        else:
            pass
        indx+=1
    plt.title(f'Average Accuracy of {cond} Hand\nFinger Decoding Across Subjects')
    plt.xlabel('Time')
    plt.ylabel('Decoding Accuracy')
    plt.ylim(y_min, y_max)
    plt.savefig(os.path.join(inputPrefixOutput, outfile))
    plt.show()
    
#print(data.head)
#Write function to plot left or right hand data
def plotBothCond(group, subject, inputPrefixOutput, cond):
    #outputDir = f'{inputPrefixOutput}GroupResults/'
    outfile = f'{inputPrefixOutput}GroupPlot_LeftVsRight_FingerDecoding_WilcoxonBonf05'
    if cond=='Left': 
        c='steelblue'
        c1 = 'lightskyblue'
        plt.figure()
    else: 
        c='mediumorchid'
        c1='violet'
    x = np.linspace(-0.5,0.5,len(subject))
    y = group['avg']

    err = group['sterr']
    err1 = []
    err2 = []
    for i in range(len(err)):
        err1.append(y[i]-err[i])
        err2.append(y[i]+err[i])
    err1 = np.array(err1).flatten()
    err2=np.array(err2).flatten()
    timepoints = len(y)
    y_min = 0.4
    y_max = 0.9
    
    plt.plot(x,y,color=c,linewidth=1)
    plt.fill_between(x, err1, err2, alpha=0.3, color=c1)
    indx = 0
    for i in group['stat_wilcoxon_Bonf']['h']:
        if group['stat_wilcoxon_Bonf']['h']==1:
            plt.scatter(indx, 0.44, color=c, s=60, zorder=5)
        else:
            pass
        indx+=1
    plt.title(f'Average Accuracy of Left and Right\nHand Finger Decoding Across Subjects')
    plt.xlabel('Time')
    plt.ylabel('Decoding Accuracy')
    plt.ylim(y_min, y_max)
    if cond=='Right':
        plt.savefig(os.path.join(inputPrefixOutput, outfile))
        plt.show()
```
